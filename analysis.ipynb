```python
import matplotlib.pyplot as plt

from zernike import ZernikeCalculator
```

```python
zern = ZernikeCalculator()
zern.plot_polynomials()
```

### Load Example Data

```python
# Load CSV data

path = "example_data.csv"

zern.load_csv(path, resolution=2048)
```

```python
plt.imshow(zern.map, cmap='inferno', interpolation='nearest')
plt.colorbar()  # Add a colorbar
plt.title('Heatmap of Heights')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```
